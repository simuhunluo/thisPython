```python
# ï¼ˆ4ï¼‰åœ¨åŽŸå§‹çš„dfä¸­æ·»åŠ ä¸€åˆ—â€˜low_destâ€™ï¼Œå†…å®¹æ˜¯â€˜destâ€™è¿™åˆ—çš„å­—ç¬¦ä¸²çš„å°å†™å½¢å¼
df['low_dest'] = df['dest'].str.lower()
print df
```

```python
# ï¼ˆ5ï¼‰ä»ŽåŽŸå§‹çš„dfä¸­é€‰å–â€˜distâ€™å¤§äºŽ1000å¹¶ä¸”å°äºŽ1200çš„æ•°æ®ä½œä¸ºä¸€ä¸ªæ–°çš„df3ï¼Œå†ä»ŽåŽŸå§‹æ•°æ®ä¸­é€‰å–â€˜timeâ€™å¤§äºŽ100æ•°æ®åšä¸ºæ–°çš„df4ï¼Œç„¶åŽå°†df3å’Œdf4ï¼ŒæŒ‰è¡Œåˆå¹¶ä¸ºä¸€ä¸ªæ•°æ®æ¡†,å‘½åä¸ºdf5
df3 = df[(df['dist'] > 1000) & (df['dist'] < 1200)]
print df3
df4 = df[df['time'] > 100]
print df4
df5 = pd.merge(df3, df4)
print df5
```

```python
# 6ï¼‰å–å‡ºdf3ä¸­çš„â€˜timeâ€™,â€˜distâ€™è¿™ä¸¤åˆ—åšä¸ºdf6ï¼Œæ±‚å‡ºdf6ä¸­æ¯åˆ—ä¸­æœ€å¤§å€¼ä¸Žæœ€å°å€¼çš„å·®å€¼ï¼ˆå¯¹æ•°æ®è¿ç”¨å‡½æ•°çš„æ–¹å¼ï¼‰
df6 = df3.loc[:, ['time', 'dist']]
print df6


def get_max_min_dif(max, min):
    return (max - min)


print get_max_min_dif(max(df6['time']), min(df6['time']))
```

```python
# ï¼ˆ7ï¼‰å°†dfæ ¹æ®â€œdestâ€åˆ†ç»„ï¼Œç»Ÿè®¡å„ç›®çš„åœ°ï¼Œéƒ½æœ‰å¤šå°‘æ¡æ•°æ®ï¼Œå‘½åä¸ºâ€œdest_countâ€ï¼Œç„¶åŽå°†â€œdestâ€å’Œâ€œdest_countâ€ç»„æˆæ–°çš„dataframeå‘½åä¸ºdf7ï¼ˆæç¤ºè®¡ç®—æ¡æ•°ç”¨np.size ï¼‰

dest_count = df.groupby(df['dest']).size()
print dest_count
df7 = pd.DataFrame(dest_count, columns=['dest_count'])
print df7
```
